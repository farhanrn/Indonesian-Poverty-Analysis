- Human Error: Data bisa hilang karena kesalahan manusia, seperti tidak sengaja menghapus file atau salah menyimpan data.

3. Data yang Disengaja Dikosongkan
  Dalam beberapa kasus, data sengaja dikosongkan untuk tujuan tertentu, seperti:
- Menjaga privasi: Data sensitif mungkin dikosongkan untuk melindungi privasi responden.
- Sebagai penanda: Nilai kosong yang sama mungkin digunakan sebagai penanda untuk data yang tidak tersedia atau tidak relevan.

Pada umumnya EDA dilakukan dengan beberapa cara:

- Multivariat Analysis  analisis yang menggunakan lebih dari atau sama dengan tiga variabel.

```python
def univariate_analysis(df):
  for column in df.columns:
    print(f"\n## Univariate Analysis for: {column}")
    print(df[column].describe())

    if pd.api.types.is_numeric_dtype(df[column]):
      # Numeric columns
      plt.figure(figsize=(12, 6))
      plt.subplot(1, 3, 1)
      plt.hist(df[column], bins=30, color='skyblue', edgecolor='black')
      plt.title('Histogram')

      plt.subplot(1, 3, 2)
      plt.boxplot(df[column], vert=True, patch_artist=True, boxprops=dict(facecolor='lightgreen'))
      plt.title('Box Plot')

      plt.subplot(1, 3, 3)
      plt.violinplot(df[column], vert=True, showmeans=True)
      plt.title('Violin Plot')

      plt.tight_layout()
      plt.show()
    elif pd.api.types.is_string_dtype(df[column]) or pd.api.types.is_object_dtype(df[column]):
      # Categorical columns
      plt.figure(figsize=(12, 6))
      value_counts = df[column].value_counts()
      bars = plt.bar(value_counts.index, value_counts.values, color='skyblue', edgecolor='black')

      plt.title('Bar Chart')
      plt.xlabel(column)
      plt.ylabel('Count')
      plt.xticks(rotation=45, ha='right')

      for bar in bars:
          yval = bar.get_height()
          plt.text(bar.get_x() + bar.get_width() / 2, yval, int(yval), ha='center', va='bottom', fontsize=10)

      plt.tight_layout()
      plt.show()
    else:
      print("Data type not recognized for visualization.")

# Assuming your DataFrame is named 'data'
univariate_analysis(data)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


def univariate_analysis(df):
  for column in df.columns:
    print(f"\n## Univariate Analysis for: {column}")
    print(df[column].describe())

    if pd.api.types.is_numeric_dtype(df[column]):
      # Numeric columns
      plt.figure(figsize=(12, 6))
      plt.subplot(1, 3, 1)
      sns.histplot(df[column], kde=True, bins=30, color='skyblue', edgecolor='black')
      plt.title('Histogram')

      plt.subplot(1, 3, 2)
      sns.boxplot(y=df[column], color='lightgreen')
      plt.title('Box Plot')

      plt.subplot(1, 3, 3)
      sns.violinplot(y=df[column], color='salmon')
      plt.title('Violin Plot')

      plt.tight_layout()
      plt.show()
    elif pd.api.types.is_string_dtype(df[column]) or pd.api.types.is_object_dtype(df[column]):
      # Categorical columns
      plt.figure(figsize=(12, 6))
      value_counts = df[column].value_counts()
      bars = plt.bar(value_counts.index, value_counts.values, color='skyblue', edgecolor='black')

      plt.title('Bar Chart')
      plt.xlabel(column)
      plt.ylabel('Count')
      plt.xticks(rotation=45, ha='right')

      for bar in bars:
          yval = bar.get_height()
          plt.text(bar.get_x() + bar.get_width() / 2, yval, int(yval), ha='center', va='bottom', fontsize=10)

      plt.tight_layout()
      plt.show()
    else:
      print("Data type not recognized for visualization.")

# Assuming your DataFrame is named 'data'
univariate_analysis(data)
```

Berdasarkan Grafik diatas, Berikut beberapa insight yang dapat diperoleh

- Grafik Density Plot with Histogram

1. **Distribusi Data**: Histogram menunjukkan distribusi data persentase penduduk miskin dengan frekuensi pada sumbu vertikal dan persentase pada sumbu horizontal. Sebagian besar data berada di kisaran 0 hingga 2000, dengan frekuensi tertinggi di sekitar 1000. Ini menunjukkan bahwa banyak daerah memiliki persentase penduduk miskin yang relatif rendah hingga menengah.

2. **Distribusi Positif Skew**: Grafik menunjukkan pola distribusi yang condong ke kanan (*positively skewed*). Ini berarti sebagian besar data terpusat di persentase rendah (sekitar 1000), sementara ada beberapa daerah yang memiliki persentase penduduk miskin yang jauh lebih tinggi (di atas 3000), meskipun frekuensinya lebih rendah.

3. **Frekuensi Puncak**: Nilai frekuensi tertinggi terjadi pada rentang 1000 hingga 1200, di mana terdapat sekitar 50 daerah yang memiliki persentase penduduk miskin dalam kisaran tersebut. Ini menunjukkan bahwa kebanyakan daerah memiliki persentase penduduk miskin yang relatif serupa, yaitu sekitar 1000-1200.

4. **Long Tail**: Bagian "ekor" di sebelah kanan histogram memperlihatkan bahwa meskipun ada beberapa daerah dengan persentase penduduk miskin yang sangat tinggi (di atas 3000 hingga lebih dari 4000), jumlah daerah tersebut jauh lebih sedikit.

- Grafik box plot
1. **Median**: Garis horizontal di dalam kotak menunjukkan nilai median. Ini adalah titik tengah dari data, yang menunjukkan bahwa separuh data berada di atas dan separuh lainnya di bawah garis ini. Nilai median pada grafik ini berada di sekitar angka 1000.

2. **Rentang Antar Kuartil (IQR)**: Kotak menggambarkan rentang antar kuartil (Q1 hingga Q3), yang mencakup 50% data. Artinya, mayoritas persentase penduduk miskin berada di antara sekitar 800 dan 2000.

3. **Outlier**: Titik-titik yang berada di atas batas garis atas adalah *outliers*, menunjukkan daerah atau populasi yang memiliki persentase penduduk miskin jauh lebih tinggi dari kelompok mayoritas. Ada beberapa data yang cukup ekstrem hingga melebihi 4000.

4. **Min dan Max (tanpa outlier)**: Garis horizontal yang berada di bawah dan di atas kotak menunjukkan batas minimum dan maksimum dari data tanpa memperhitungkan outliers. Nilai minimum terlihat mendekati 0, sedangkan nilai maksimum berada di kisaran 3000-an.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Group data by province and calculate the average persentase_penduduk_miskin
province_poverty = data.groupby('provinsi')['persentase_penduduk_miskin'].mean()

# Create a bar chart with different colors based on the average poverty rate
plt.figure(figsize=(15, 8))
bars = plt.bar(province_poverty.index, province_poverty.values, color='skyblue')

# Define color thresholds for highlighting
low_threshold = province_poverty.quantile(0.25)  # Lower quartile
high_threshold = province_poverty.quantile(0.75)  # Upper quartile

# Highlight provinces based on their poverty rate
for i, bar in enumerate(bars):
    if province_poverty.values[i] < low_threshold:
        bar.set_color('lightgreen')  # Low poverty rate
    elif province_poverty.values[i] > high_threshold:
        bar.set_color('salmon')  # High poverty rate

    # Add data label to each bar
    plt.text(
        bar.get_x() + bar.get_width() / 2,  # X-coordinate
        bar.get_height(),  # Y-coordinate (height of the bar)
        f'{bar.get_height():.2f}',  # Label with 2 decimal places
        ha='center',  # Horizontal alignment
        va='bottom',  # Vertical alignment
        fontsize=10  # Font size
    )

# Add labels and titles
plt.title('Average Poverty Rate by Province in Indonesia', fontsize=16)
plt.xlabel('Province', fontsize=14)
plt.ylabel('Average Poverty Rate (%)', fontsize=14)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```

Terlihat bahwa rata-rata penduduk miskin terbesar berada pada provinsi Papua sebesar 2497 dan maluku 2144

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.scatterplot(x='partisipasi_angkatan_kerja', y='tingkat_pengangguran_terbuka', hue='klasifikasi_kemiskinan', data=data)
plt.title('Hubungan Partisipasi Angkatan Kerja dengan Tingkat Pengangguran Terbuka dan Klasifikasi Kemiskinan')
plt.xlabel('Partisipasi Angkatan Kerja')
plt.ylabel('Tingkat Pengangguran Terbuka')

handles, labels = plt.gca().get_legend_handles_labels()
new_labels = ['Tidak Miskin', 'Miskin']
plt.legend(handles=handles, labels=new_labels, title='Klasifikasi Kemiskinan')

plt.show()
```

1. **Hubungan Negatif antara Partisipasi Angkatan Kerja dan Tingkat Pengangguran Terbuka:**

2. **Perbedaan antara Miskin dan Tidak Miskin:**
   - Titik **oranye** (miskin) lebih sering terlihat di wilayah dengan **tingkat pengangguran terbuka yang lebih tinggi** dan **partisipasi angkatan kerja yang lebih rendah**. Ini menunjukkan bahwa di wilayah-wilayah miskin, partisipasi angkatan kerja cenderung lebih rendah, sementara tingkat pengangguran lebih tinggi.
   - Titik **biru** (tidak miskin) lebih terdistribusi pada wilayah dengan **tingkat partisipasi angkatan kerja yang lebih tinggi** dan **tingkat pengangguran yang lebih rendah**, menegaskan bahwa wilayah yang tidak tergolong miskin cenderung memiliki kondisi ekonomi yang lebih baik.

3. **Distribusi Titik Miskin (Oranye):**
   - Wilayah miskin cenderung berkumpul di area dengan partisipasi angkatan kerja rendah (sekitar 6000 hingga 7000) dan pengangguran yang lebih tinggi (sekitar 400 hingga 1200). Ini menandakan bahwa pengangguran yang lebih tinggi sering terjadi di daerah-daerah dengan status kemiskinan, yang mungkin berhubungan dengan peluang kerja yang lebih sedikit atau pendidikan yang kurang memadai.

4. **Outlier:**
   - Beberapa titik oranye terlihat memiliki partisipasi angkatan kerja yang tinggi (di atas 8000) namun masih tergolong miskin. Hal ini bisa menunjukkan bahwa walaupun banyak orang yang berpartisipasi dalam angkatan kerja, kualitas atau jenis pekerjaan yang tersedia mungkin tidak memadai untuk mengangkat mereka keluar dari kemiskinan.

- **Kemiskinan** berkaitan dengan **tingkat pengangguran yang lebih tinggi** dan **partisipasi yang lebih rendah**, tetapi ada juga beberapa daerah miskin yang memiliki tingkat partisipasi yang tinggi, mengindikasikan adanya faktor-faktor lain seperti kualitas pekerjaan yang perlu dipertimbangkan.

# Kesimpulan dan Future Action 

Meningkatkan kualitas pendidikan dan keterampilan adalah langkah penting dalam upaya pengentasan kemiskinan dan peningkatan kesejahteraan masyarakat. Pendidikan yang berkualitas, disertai dengan keterampilan yang relevan, dapat membuka peluang kerja yang lebih baik bagi individu, sehingga membantu mereka keluar dari jeratan kemiskinan. Penelitian menunjukkan bahwa terdapat hubungan erat antara tingkat pendidikan yang rendah dengan tingkat kemiskinan yang tinggi.

## Strategi untuk Meningkatkan Kualitas Pendidikan dan Keterampilan

**1. Memperluas Akses Pendidikan**
- Fokus pada peningkatan akses pendidikan di daerah pedesaan dan terpencil untuk memastikan bahwa semua anak, tanpa memandang lokasi, memiliki kesempatan untuk mendapatkan pendidikan yang layak.

**2. Meningkatkan Kualitas Guru dan Kurikulum**
- Pelatihan berkelanjutan bagi guru sangat penting untuk memperbarui keterampilan mengajar mereka. Program sertifikasi dan pengembangan kurikulum yang relevan dengan kebutuhan siswa dapat meningkatkan kualitas pengajaran.
- Penggunaan teknologi dalam pembelajaran juga harus diperkenalkan agar guru dapat mengelola kelas dengan lebih efektif dan menarik.

**3. Menyediakan Program Pelatihan Vokasi**
- Program pelatihan vokasi yang sesuai dengan kebutuhan pasar kerja harus diperluas. Ini akan membantu siswa memperoleh keterampilan praktis yang langsung dapat diaplikasikan di dunia kerja.

**4. Memberikan Beasiswa dan Bantuan Pendidikan**
- Beasiswa dan bantuan pendidikan perlu disediakan untuk siswa dari keluarga miskin agar mereka dapat melanjutkan pendidikan tanpa beban finansial yang berat.

## Menciptakan Lapangan Kerja

Penciptaan lapangan kerja yang memadai dan berkualitas sangat penting untuk meningkatkan pendapatan masyarakat serta mengurangi angka pengangguran, terutama di kalangan pemuda.

**Langkah-langkah yang Dapat Diambil:**
- **Mendorong Investasi Ekonomi:** Pemerintah perlu menciptakan iklim investasi yang kondusif untuk menarik investor, baik domestik maupun asing.
- **Pengembangan UMKM:** Sektor usaha mikro, kecil, dan menengah (UMKM) harus didorong melalui pelatihan dan akses ke modal.
- **Insentif bagi Perusahaan:** Memberikan insentif kepada perusahaan yang berkomitmen menciptakan lapangan kerja baru.
- **Peningkatan Kualitas Tenaga Kerja:** Melalui pelatihan dan pendidikan, kualitas tenaga kerja dapat ditingkatkan sehingga lebih siap menghadapi tuntutan pasar.

## Memberikan Bantuan Sosial Tepat Sasaran

Bantuan sosial merupakan alat penting dalam membantu masyarakat miskin memenuhi kebutuhan dasar mereka.

**Strategi untuk Meningkatkan Efektivitas Bantuan Sosial:**
- **Perbaikan Sistem Data Terpadu:** Memastikan bahwa data penerima bantuan sosial akurat agar bantuan tepat sasaran.
- **Efektivitas Program:** Meningkatkan efektivitas program bantuan sosial seperti Program Keluarga Harapan (PKH) dan Bantuan Pangan Non Tunai (BPNT).
- **Inovasi dalam Program Bantuan Sosial:** Mengembangkan program-program baru yang berkelanjutan untuk menjangkau lebih banyak orang.

## Meningkatkan Akses terhadap Layanan Kesehatan

Kesehatan yang baik adalah kunci untuk meningkatkan produktivitas masyarakat. Oleh karena itu, akses terhadap layanan kesehatan perlu diperluas.

**Langkah-langkah Strategis:**
- **Perluasan Jaringan Fasilitas Kesehatan:** Khususnya di daerah pedesaan, fasilitas kesehatan harus ditingkatkan.
- **Peningkatan Kualitas Layanan Kesehatan:** Fokus pada peningkatan kualitas layanan agar masyarakat mendapatkan perawatan yang memadai.
- **Pengurangan Biaya Layanan Kesehatan:** Upaya untuk menurunkan biaya layanan kesehatan agar lebih terjangkau bagi masyarakat.
- **Promosi Gaya Hidup Sehat:** Mendorong masyarakat untuk menerapkan pola hidup sehat sebagai langkah preventif terhadap penyakit.

## Memperkuat Infrastruktur Dasar

Infrastruktur dasar yang memadai sangat penting untuk mendukung aksesibilitas dan produktivitas masyarakat.

**Langkah-langkah Pembangunan Infrastruktur:**
- **Pembangunan Jalan dan Jembatan:** Memperbaiki infrastruktur transportasi untuk meningkatkan konektivitas antar daerah.
- **Akses Listrik dan Air Bersih:** Memastikan bahwa semua wilayah, terutama pedesaan, memiliki akses terhadap listrik dan air bersih.
- **Pengembangan Infrastruktur Telekomunikasi:** Mengembangkan jaringan telekomunikasi untuk mendukung komunikasi dan informasi.

Rekomendasi solusi ini bersifat umum namun memberikan kerangka kerja bagi pemerintah Indonesia dalam upaya pengentasan kemiskinan. Untuk mencapai hasil yang lebih spesifik dan efektif, analisis data mendalam serta keterlibatan berbagai pemangku kepentingan sangat diperlukan dalam perumusan dan implementasi solusi tersebut.
